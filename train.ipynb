```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import os
```

```python
df = pd.read_csv('dataset/final_df.csv')
df
```

```python
df.set_index('Brats20ID', inplace=True)
df
```

```python
# data type should be int but it is of type object, hence there is an outlier present
df['Survival_days']
```

```python
df[df['Survival_days'].apply(lambda x: str(x).isdigit() == False)]
```

```python
idx = df.index[df['Survival_days'].apply(lambda x: str(x).isdigit() == False)]
idx
```

```python
df.at[idx[0], 'Survival_days']
```

```python
df.drop(idx, inplace=True)
```

```python
df['Survival_days'] = df['Survival_days'].astype(int)
```

```python
df.dtypes.value_counts()
```

```python
df = pd.get_dummies(df, columns=['Extent_of_Resection'])
df
```

```python
corr = df.corr()
target_corr = abs(corr['Survival_days'])
relevant_features = target_corr[target_corr > 0.15]
cols = relevant_features.index
new_df = df[cols]
```
