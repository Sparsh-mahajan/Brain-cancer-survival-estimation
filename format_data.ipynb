```python
import pandas as pd
import numpy as np
import os
```

```python
final_df = pd.DataFrame()
df_dir = 'dataset'

# for each label, read in the csv file and add the label to the column names
for label in os.listdir(df_dir):
    for file in os.listdir(os.path.join(df_dir, label)):
        df = pd.read_csv(os.path.join(df_dir, label, file))
        # set the index to the Brats20ID
        df.set_index(['Brats20ID'], inplace = True, append = False, drop = True)
        # add the label to the column names
        new_cols = []
        for col in df.columns:
            new_cols.append(label + '_' + col)
        df.columns = new_cols
        # add the dataframe to the final dataframe
        final_df = pd.concat([final_df, df], axis=1, join='outer', ignore_index=False, sort=False)
final_df.info()
```

```python
# read in the survival info csv
survival_df = pd.read_csv('MICCAI_BraTS2020_TrainingData\survival_info.csv')
```

```python
final_df.info()
```

```python
# Number of rows with NaN values
final_df.shape[0] - final_df.dropna().shape[0]
```

```python
# drop rows with NaN values
final_df.dropna(inplace=True)

# remove unnecessary columns (A column with the index is added when reading in the csv, i.e. 'Unnamed: 0')
drop_cols = []
for col in final_df.columns:
    if col.endswith('Unnamed: 0'):
        drop_cols.append(col)
final_df.drop(drop_cols, axis=1, inplace=True)
final_df.info()
```

```python
final_df
```

```python
# merge the survival info with the final dataframe
final_df = final_df.merge(survival_df, how='inner', left_index=True, right_on='Brats20ID')
```

```python
# set the index to the Brats20ID
final_df.set_index(['Brats20ID'], inplace=True, append=False, drop=True)
final_df
```

```python
# save the final dataframe to a csv file
final_df.to_csv('dataset/final_df.csv')
```
